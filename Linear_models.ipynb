### Gradient Descent
**Gradient w ve b ni deyiserek en ideal varianti tapir(en asagi loss)**  


Gradient Descent alqoritmi, modelin w (çəkilər) və b (bias) kimi parametrlərini təkrarlı şəkildə yeniləyərək zərər funksiyasını (loss function) minimuma endirməyə çalışır.


Gradient Descent alqoritminin əsas məqsədi theta best-i (ən optimal parametrləri) tapmaqdır. Bu, modelin zərər funksiyasını (loss function) minimuma endirərək edilir.

 **Qlobal Minimum:**

Tərif: Qlobal minimum, funksiyanın ən aşağı dəyərinə çatdığı nöqtədir. Yəni, bütün mümkün parametr dəyərləri arasında ən kiçik zərər dəyərinə malik olan nöqtədir.

Şərh: Qrafikdə qlobal minimum, ən aşağı nöqtə kimi görünə bilər. Bu nöqtə, modelin ən yaxşı performans göstərdiyi və ən aşağı xəta dərəcəsinə sahib olduğu yerdir. Gradient descent üsullarının məqsədi bu nöqtəyə çatmaqdır.

 **Lokal Minimum:**

Tərif: Lokal minimum, funksiyanın müəyyən bir bölgədə ən aşağı dəyərinə çatdığı, lakin qlobal minimumdan daha yüksək bir dəyərə sahib olduğu nöqtədir. Yəni, bu nöqtə yaxın ətrafdakı ən aşağı nöqtə ola bilər, lakin bütün funksiya üzrə ən aşağı nöqtə deyil.

Şərh: Qrafikdə lokal minimum, bir çuxur və ya aşağı bir nöqtə kimi görünə bilər, lakin bu nöqtə qlobal minimum qədər dərin deyil. Model bu nöqtədə ilişib qala bilər və daha yaxşı bir həll tapa bilməz."

1. **Batch Gradient Descent (BGD):**
* Bütün verilənləri eyni anda istifadə edir.
* Yavaşdır və çox böyük verilənlər üçün resurs tələb edir.
* Sabit və dəqiqdir, amma lokal minimumda ilişə bilər.

2. **Stochastic Gradient Descent (SGD):**
* Hər dəfə bir nümunə istifadə edir.
* Sürətlidir və böyük verilənlər üçün uyğundur.
* Yüksək dəyişkənlik göstərir və optimal nöqtəyə tam yaxınlaşmaya bilər, amma local minimumlardan çıxa bilər.

3. **Mini-Batch Gradient Descent:**
* Kiçik partiyalarla işləyir.
* BGD ilə müqayisədə daha sürətlidir və SGD-nin dəyişkənliyini azaldır.
Daha balanslıdır, həm sürətli, həm də daha stabil nəticələr verir.

Sabit nəticə: Modelin itki funksiyası (loss function) təlim zamanı çox dəyişməməlidir, yəni hər iterasiyada fərqli nəticələr əldə edilməməlidir.


**Stochastic Gradient Descent (SGD) kimi üsullar lokal minimumlardan qaçmaq baxımından daha uğurlu ola bilər.**

 **Batch Gradient Descent kimi üsullar daha sabit, lakin daha yavaş olduqları üçün lokal minimumlarda ilişib qalma riski daha yüksək ola bilər.**

**Mini-batch Gradient Descent həm SGD-nin lokal minimumlardan qaçmaq qabiliyyətini, həm də Batch GD-nin sabitliyini birləşdirir.**

**Learning rate** (öyrənmə dərəcəsi), maşın öyrənmə alqoritmalarında modelin nə qədər sürətlə öyrənəcəyini müəyyən edən bir parametrdir. Gradient Descent kimi optimallaşdırma alqoritmalarında learning rate, hər iterasiyada modelin parametrlərinin nə qədər dəyişəcəyini göstərir.

Yüksək learning rate daha sürətli öyrənmə təmin edir, lakin optimal nöqtəni qaçırma riski var.
Aşağı learning rate daha yavaş, amma sabit bir öyrənmə təklif edir.
Doğru learning rate seçimi modelin nəticələrini və təlim müddətini təsir edir.

 Batch(or Full) Gradient

```python
eta= 0.1 #learing rate
n_epochs= 1000
m= len(X_b) # number of instance

np.random.seed(42)
theta= np.random.randn(2,1) # randomly initialize model parameters

for epoch in range(n_epochs):
  gradient= 1/m*2 *X_b.T @(X_b @ theta- y)
  theta=theta-eta*gradient
```

Stochastic Gradient Descend

```python
np.random.seed(42)
eta= 0.1
m= len(X_b)
theta= np.random.randn(2,1)

for epoch in range(n_epochs):
  for iteration in range(m):
    random_index= np.random.randint(m)
    xi=X_b[random_index : random_index+1] # matrix kimi qaytrasin deye
    yi=y[random_index : random_index+1]
    gradients= 2*xi.T @ (xi @ theta-yi)
    theta= theta- eta*gradients
```

  Mini Batch Gradient Descend

```python
np.random.permutation(m)
```

```python
from math import ceil
n_epochs= 50
eta= 0.1
m=len(X_b)
minibatch_size=20
n_batches_per_epoch= ceil(m/minibatch_size)

np.random.seed(42)
theta=np.random.randn(2,1)

for epoch in range(n_epochs):
  shuffled_indices= np.random.permutation(m)
  X_b_shuffled= X_b[shuffled_indices]
  y_shuffled= y[shuffled_indices]
  for iteration in range(0, n_batches_per_epoch):
    idx=iteration * minibatch_size
    xi= X_b_shuffled[idx: idx+minibatch_size]
    yi= y_shuffled[idx: idx+minibatch_size]
    gradients= 2/ minibatch_size * xi.T @ (xi @ theta - yi)
    theta=theta-eta*gradients
```
